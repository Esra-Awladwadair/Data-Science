```python
import yfinance as yf

# Get Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="1y")

# Reset the index
tesla_data.reset_index(inplace=True)

# Save the DataFrame to a CSV file (optional)
tesla_data.to_csv('tesla_stock_data.csv', index=False)

# Display the first five rows
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Get Tesla revenue data from Yahoo Finance
url = "https://finance.yahoo.com/quote/TSLA/financials?p=TSLA"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extract revenue data
revenue_data = []
rows = soup.find_all('tr')

# Find the row with the revenue data
for row in rows:
    if 'Total Revenue' in row.text:
        cols = row.find_all('td')
        revenue_data = [col.text for col in cols[1:]]  # Skip the first column (label)
        break

# Convert to DataFrame for easier handling
tesla_revenue = pd.DataFrame(revenue_data, columns=['Revenue'])

# Display the last five rows
print(tesla_revenue.tail())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Get GameStop revenue data from Yahoo Finance
url = "https://finance.yahoo.com/quote/GME/financials?p=GME"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extract revenue data
revenue_data = []
rows = soup.find_all('tr')

# Find the row with the revenue data
for row in rows:
    if 'Total Revenue' in row.text:
        cols = row.find_all('td')
        revenue_data = [col.text for col in cols[1:]]  # Skip the first column (label)
        break

# Convert to DataFrame for easier handling
gme_revenue = pd.DataFrame(revenue_data, columns=['Revenue'])

# Display the last five rows
print(gme_revenue.tail())
```

```python
import matplotlib.pyplot as plt

# Define the make_graph function
def make_graph(data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data['Date'], data['Close'], label='Tesla Stock Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price ($)')
    plt.legend()
    plt.grid()
    plt.show()

# Call the make_graph function with Tesla data
make_graph(tesla_data, 'Tesla Stock Price Over the Last Year')
```
